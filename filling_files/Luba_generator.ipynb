### Author

```python
from mimesis.enums import Gender
from mimesis import Generic as gnr

def get_author(n:int, actual_year = 2022):
    """generator of author"""
    person = mim.Person('ru')
    d = gnr('ru')
    min_age, max_age = 16, 80
    
    for i in  range(n):
        birthday = d.datetime.date(start = actual_year - max_age, end = actual_year - min_age)
        sex = person.sex()[:1]
        if (sex == "М"):
            gender = 'male'
        elif (sex == "Ж"):
            gender = 'female'
        else:
            continue
        
        yield {
            'name': person.full_name(Gender(gender)), 
            'sex': sex, 
            'num_of_works': int(min(60, np.random.exponential(5))) + 1,
            'birthday': birthday,
            'age': actual_year - birthday.year
        }

for i in get_author(10):
    print(i)
```

```python
count_authors = 100

AUTHORS = pd.DataFrame([x for x in get_author(count_authors)])
AUTHORS.to_csv('AUTHORS.csv', header=False)
AUTHORS
```

### Works

```python
# WORKS
from mimesis import Text

def Title():
    text=[]
    while not text:
        edge = randint(1, 4)
        bias = randint(0, 3)
        text = Text('ru').text(1).split(' ')
        text = text[edge : min(1 + edge + bias, len(text))]
    
    text[0] = text[0].title()
    text = " ".join(text)
    return text

Title()
```

```python
from mimesis import Generic as gnr

work_count = 1000

def get_work(n:int):
    
    d = gnr('ru')
    for i in range(n):
        author_id = randint(0, count_authors-1)
        birthday_year = AUTHORS.iloc[author_id].birthday.year
        age = AUTHORS.iloc[author_id].age
        
        yield {
            "work_name": Title(),
            'work_author_id': author_id,
            'work_creating_date': d.datetime.date(start=birthday_year, end=birthday_year + age),
            'work_number_of_words': int(min(1e6, np.random.exponential(1000))) + 1000
        }
        
for work in get_work(1):
    print(AUTHORS.iloc[work['work_author_id']])
    print(work)

WORKS = pd.DataFrame([x for x in get_work(work_count)])
WORKS
```

```python
books_per_author = WORKS.groupby(['work_author_id']).size()
AUTHORS.num_of_works = books_per_author
```

```python
# wrtie this
WORKS.to_csv('WORKS.csv', header=False)
AUTHORS.to_csv('AUTHORS.csv', header=False)
```

### Shops

```python
# SHOPS

from mimesis import Address
from random import getrandbits

def get_shop(n:int):
    person = mim.Person('ru')
    addresser = Address('ru')
    for _ in range(n):
        address = addresser.address()
        name = person.name(Gender('female'))
        
        yield {
            'shop_name': name,
            'shop_adress': address,
            'shop_web_store': getrandbits(1)
        }
```

```python
shop_count = 100

SHOPS = pd.DataFrame([x for x in get_shop(shop_count)])
SHOPS
```

```python
SHOPS.to_csv('SHOPS.csv', header=False)
```

### Publishing Houses

```python
# PUBLISHING_HOUSES
from mimesis.enums import Gender
from mimesis import Generic as gnr


def get_publisher(n:int):
    person = mim.Person('ru')
    d = gnr('ru')
    for _ in range(n):
        yield {
            'house_name': person.name(Gender('male')), 
            'house_opening_date': d.datetime.date(start=1890, end=2022),
            'house_number_of_books': randint(0, 1000)
        }
```

```python
publisher_count = 15

PUBLISHING_HOUSES = pd.DataFrame([x for x in get_publisher(publisher_count)])
PUBLISHING_HOUSES
```

### Books

```python
# BOOKS

def get_book(n:int):
    d = gnr('ru')
    for _ in range(n):
        publisher_idx = randint(0, publisher_count - 1)
        open_year = PUBLISHING_HOUSES.iloc[publisher_idx].house_opening_date.year
        yield {
            'book_shop_id': randint(0, shop_count - 1),
            'book_work_id': randint(0, work_count - 1),
            'book_house_id': publisher_idx,
            'book_publishing_date': d.datetime.date(start=open_year, end=2022),
            'book_price': max(0, int(np.random.normal(400, 100))),
            'book_hard': getrandbits(1),
            'book_in_store': max(0, int(np.random.normal(250, 100))),
            'book_rating': randint(1, 10)
        }
```

```python
books_count = int(1e3)

BOOKS = pd.DataFrame([x for x in  get_book(books_count)])
BOOKS
```

```python
# update num of books in publisher

books_per_publisher = BOOKS.groupby(['book_house_id']).size()
PUBLISHING_HOUSES.house_number_of_books = books_per_publisher
```

```python
PUBLISHING_HOUSES.to_csv('PUBLISHING_HOUSES.csv', header=False)
BOOKS.to_csv('BOOKS.csv', header=False)
```
